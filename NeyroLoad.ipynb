```
from PIL import Image
import numpy as np
import glob
import matplotlib.pyplot as plt

from keras.models import Sequential,load_model
from keras.layers import Dense, Dropout,Activation, Flatten,MaxPooling2D,Conv2D
from keras.utils import np_utils
from keras import metrics
import tensorflow as tf
from tensorflow.python.client import device_lib
from keras import backend as bac
bac.tensorflow_backend._get_available_gpus()
from matplotlib import pyplot as plt
```

```
im_size=(100,100)

def ImageArray(adress1,adress2):
  size = (len(adress1) + len(adress2))*2
  print(size)
  allsize=(size,100,100,3)
  onesize=(100,100,3)
  im_data = np.zeros(allsize)
  metka = np.zeros(size)
  schet=0
  for name in adress1:
    print(schet)
    #print(name)
    im=Image.open(name)
    im = im.resize(im_size)
    if (np.asarray(im).shape!=onesize):
      print(name)
    else:
      im_data[schet]=np.asarray(im,order='F')
      metka[schet]=0
      schet=schet+1
      im_data[schet]=np.asarray(im.transpose(Image.FLIP_LEFT_RIGHT),order='F')
      metka[schet]=0
      schet=schet+1

  for name in adress2:
    print(schet)
    #print(name)
    im=Image.open(name)
    im = im.resize(im_size)
    if (np.asarray(im).shape!=onesize):
      print(name)
    else:
      im_data[schet]=np.asarray(im,order='F')
      metka[schet]=1
      schet=schet+1
      im_data[schet]=np.asarray(im.transpose(Image.FLIP_LEFT_RIGHT),order='F')
      metka[schet]=1
      schet=schet+1

  im_data=im_data.astype('float32')
  im_data /= 255
  return im_data, metka, schet
 
```

```
sl2=glob.glob('drive/My Drive/Neyr1/Sail_Test/*.jpg')
st2=glob.glob('drive/My Drive/Neyr1/Steam_Test/*.jpg')
Test_im,Test_met,testsize=ImageArray(sl2,st2)
```

```
model = load_model('drive/My Drive/Neyr1/Neyro1_model1.h5')
```

```
score = model.evaluate(Test_im,Test_met)
print(score)
```

```
con_mat = tf.math.confusion_matrix(model.predict_classes(Test_im), Test_met, 
                                   num_classes=2, dtype=tf.dtypes.int32)
print(bac.eval(con_mat))
```
